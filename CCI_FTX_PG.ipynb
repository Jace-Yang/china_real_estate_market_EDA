````python
OLD_COLUMN = ['OBJ_TIT', 'OBJ_OTHNAME', 'OBJ_URL', 'OBJ_PATH', 'OBJ_TEL', 'OBJ_SCORE', 'OBJ_PRICE', 
         'OBJ_MAJORTYPE', 'OBJ_ADD', 'OBJ_TAG', 'OBJ_OPENDATE']
NEW_COLUMN = ['OBJ_ADD', 'OBJ_MAJORTYPE', 'OBJ_NEWS_TIT', 'OBJ_NEWS_URL', 'OBJ_OPENDATE', 'OBJ_OTHNAME', 'OBJ_PATH', 'OBJ_POSTS_NUM', 'OBJ_PRICE', 'OBJ_SCORE', 'OBJ_TAG', 'OBJ_TIT', 'OBJ_URL']
print("已删除字段: ", set(OLD_COLUMN) - set(NEW_COLUMN))
print("新增字段: ", set(NEW_COLUMN) - set(OLD_COLUMN))
````

````python
fin = open(file_firstpage, 'r', encoding='utf-8')
lines = fin.readlines()
fin.close()
datalist = []


counter = 0
GET_COLUMN = ['OBJ_TIT', 'OBJ_OTHNAME', 'OBJ_URL', 'OBJ_PATH', 'OBJ_SCORE', 'OBJ_PRICE', 
         'OBJ_MAJORTYPE', 'OBJ_ADD', 'OBJ_TAG', 'OBJ_OPENDATE',
             'OBJ_POSTS_NUM', 'OBJ_NEWS_TIT', 'OBJ_NEWS_URL']

ALL_COLUMN = ["OBJ_NUM"] + GET_COLUMN
for line in lines:
    if line[0] == ',':
        line = line[1:]
    try:
        obj = json.loads(line, strict=False)
        counter += 1
        if counter ==1 :
            print(obj.keys())        
        datalist.append([str(counter)] + [obj[col] for col in GET_COLUMN])
    except:
        print(counter)
        continue

df_obj_1 = pd.DataFrame(datalist, columns = ALL_COLUMN).drop_duplicates(subset=['OBJ_TIT', 'OBJ_URL'])
del datalist
df_obj_1
````

````python
fin = open(file_detail, 'r', encoding='utf-8-sig')
lines = fin.readlines()
fin.close()
datalist = []
i = 0
for line in lines:
    if line[0] == ',': 
        line = line[1:]
    i += 1
    try:
        obj = json.loads(line, strict=False)
        OBJ_URL = obj['OBJ_URL']
        OBJ_ESTYPE = obj['OBJ_ESTYPE']
        OBJ_CONTYPE = obj['OBJ_CONTYPE']
        OBJ_PROPYEAR = obj['OBJ_PROPYEAR']
        OBJ_DEVELOPER = obj['OBJ_DEVELOPER']
        OBJ_FEATURE = obj['OBJ_FEATURE']
        OBJ_DECORATION = obj['OBJ_DECORATION']
        OBJ_LOOPLOCATION = obj['OBJ_LOOPLOCATION']
        OBJ_SALESTATE = obj['OBJ_SALESTATE']
        OBJ_SALEDISCOUNT = obj['OBJ_SALEDISCOUNT']
        OBJ_DELIDATE = obj['OBJ_DELIDATE']
        OBJ_LAYCOUNT = obj['OBJ_LAYCOUNT']
        OBJ_SURROUNDIFO = obj['OBJ_SURROUNDIFO']
        OBJ_TOTLANDAREA = obj['OBJ_TOTLANDAREA']
        OBJ_FAR = obj['OBJ_FAR']
        OBJ_PARKINGNUM = obj['OBJ_PARKINGNUM']
        OBJ_HOUSENUM = obj['OBJ_HOUSENUM']
        OBJ_ESCOST = obj['OBJ_ESCOST']
        OBJ_ESDECRIP = obj['OBJ_ESDECRIP']
        OBJ_FLOORINFO = obj['OBJ_FLOORINFO']
        OBJ_TOTCONAREA = obj['OBJ_TOTCONAREA']
        OBJ_GR = obj['OBJ_GR']
        OBJ_BUILDINGNUM = obj['OBJ_BUILDINGNUM']
        OBJ_ESCOMP = obj['OBJ_ESCOMP']
        OBJ_INTRO = obj['OBJ_INTRO']
        datalist.append([OBJ_URL, OBJ_ESTYPE, OBJ_CONTYPE, OBJ_PROPYEAR, OBJ_DEVELOPER, OBJ_FEATURE, OBJ_DECORATION, OBJ_LOOPLOCATION, OBJ_SALESTATE, OBJ_SALEDISCOUNT,
                        OBJ_DELIDATE, OBJ_LAYCOUNT, OBJ_SURROUNDIFO, OBJ_TOTLANDAREA, OBJ_FAR, OBJ_PARKINGNUM, OBJ_HOUSENUM, OBJ_ESCOST, OBJ_ESDECRIP,
                        OBJ_FLOORINFO, OBJ_TOTCONAREA, OBJ_GR, OBJ_BUILDINGNUM, OBJ_ESCOMP, OBJ_INTRO])
    except:
        print(i)
        continue

columns = ['OBJ_URL', 'OBJ_ESTYPE', 'OBJ_CONTYPE', 'OBJ_PROPYEAR', 'OBJ_DEVELOPER', 'OBJ_FEATURE', 'OBJ_DECORATION', 'OBJ_LOOPLOCATION', 'OBJ_SALESTATE', 'OBJ_SALEDISCOUNT',
           'OBJ_DELIDATE', 'OBJ_LAYCOUNT', 'OBJ_SURROUNDIFO', 'OBJ_TOTLANDAREA', 'OBJ_FAR', 'OBJ_PARKINGNUM', 'OBJ_HOUSENUM', 'OBJ_ESCOST', 'OBJ_ESDECRIP',
           'OBJ_FLOORINFO', 'OBJ_TOTCONAREA', 'OBJ_GR', 'OBJ_BUILDINGNUM', 'OBJ_ESCOMP', 'OBJ_INTRO']
df_obj_3 = pd.DataFrame(datalist, columns=columns).drop_duplicates()
df_result = pd.merge(df_obj, df_obj_3, how='left', on=['OBJ_URL'])
del datalist

df_result = df_result.drop(columns=['REV_TYPE', 'REV_INTRO', 'REV_TAG', 'replys'])
df_result
````

````python
# 统计缺失情况
def report_miss(df):
    print(df.isnull().sum())
    return df

def get_address(df, column, cut = True, drop_dizhi = True):
    '''
    :cut: 如果cut的话就只留前三列
    '''
    
    if cut:
        df = df[["被参控公司","搜索公司","搜索公司地址"]].reset_index(drop = True)  
        
    target = df.loc[:,column].tolist()
    address = cpca.transform(target, pos_sensitive=True)
    
    if drop_dizhi:
        address = address.drop('地址', 1)
        
    return pd.concat([df,address], axis = 1)

def get_miss(df):
    print(df.info())
    return df

def read_pickle_file(file):
    pickle_data = pd.read_pickle(file)
    return pickle_data
````
